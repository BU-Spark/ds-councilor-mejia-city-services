```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import warnings

warnings.filterwarnings("ignore")

RRF = pd.read_csv('RRF_funds.csv')
RRF = RRF.dropna(axis=1, how='all')
RRF = RRF.drop(columns=['Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25', 'Unnamed: 26', 'Unnamed: 27', 'Unnamed: 28'])
RRF
```

```python
Approved = RRF.loc[RRF['Disposition of case'] == 'Approved']
```

```python
Approved.dropna(subset=["Race and Ethnicity"],inplace=True)
```

```python
Approved.dropna(subset=["Applicant Zip Code"],inplace=True)
```

```python
from pyproj import Transformer
import folium
from folium.plugins import HeatMap
```

```python
Approved["Applicant Zip Code"] = Approved["Applicant Zip Code"].astype('int64')
Approved["Applicant Zip Code"] = Approved["Applicant Zip Code"].astype('str')
Approved["Applicant Zip Code"] = "0" + Approved["Applicant Zip Code"]
```

```python
import pgeocode

nomi = pgeocode.Nominatim('us')
count = len(Approved)
Approved["longitude"] = np.nan
Approved["latitude"] = np.nan
Approved.reset_index(drop=True, inplace=True)

for i in range(count):
    Approved.at[i, "longitude"] = nomi.query_postal_code(Approved.at[i, "Applicant Zip Code"]).longitude
    Approved.at[i, "latitude"] = nomi.query_postal_code(Approved.at[i, "Applicant Zip Code"]).latitude
```

```python
Approved.to_csv('Approved_update.csv', index=False)
```

```python
Approved = Approved.loc[Approved['Applicant Zip Code'] != '02']
```

```python
counts = Approved['Applicant Zip Code'].value_counts().to_frame()
counts.rename(columns={'Applicant Zip Code':'count'}, inplace = True)
counts["longitude"] = np.nan
counts["latitude"] = np.nan
counts
```

```python
count_list = []
for index, row in counts.iterrows():
    counts['longitude'][index] = nomi.query_postal_code(index).longitude
    counts['latitude'][index] = nomi.query_postal_code(index).latitude
    if index != "02104":
        count_list.append([counts['latitude'][index], counts['longitude'][index], float(counts['count'][index])])
    else:
        count_list.append([42.3389, -70.9196, float(counts['count'][index])])
m = folium.Map(location = [42.3601, -71.0589],
               zoom_start = 12)
HeatMap(count_list).add_to(folium.FeatureGroup(name='Heat Map').add_to(m))
m.save('Approved_heatmap.html')
```
